```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:\ALLMLPROJ\meta-overfit-predictor\meta_dataset\meta_features.csv")

print(df.head())
print(df["overfit_label"].value_counts())
```

```python
features_to_plot = [
    "gap_slope",
    "train_acc_slope",
    "val_acc_slope",
    "train_loss_slope",
    "val_loss_slope"
]

for feat in features_to_plot:
    plt.figure()
    sns.boxplot(x="overfit_label", y=feat, data=df)
    plt.title(f"{feat} vs overfit_label")
    plt.show()
```

```python
sns.pairplot(
    df,
    vars=[
        "gap_slope",
        "train_acc_slope",
        "val_acc_slope",
        "early_train_acc",
        "early_val_acc"
    ],
    hue="overfit_label",
    diag_kind="kde"
)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(
    df.drop(columns=["run_name", "overfit_label"]).corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Feature Correlation Heatmap")
plt.show()
```
